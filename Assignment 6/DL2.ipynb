```python
# Xây dựng mô hình Neural Network
nn_model = Sequential()
nn_model.add(Dense(32, activation='relu', input_dim=X_train.shape[1]))
nn_model.add(Dropout(0.2))
nn_model.add(Dense(16, activation='relu'))
nn_model.add(Dropout(0.2))
nn_model.add(Dense(1, activation='sigmoid'))

# Compile mô hình Neural Network
nn_model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])

# In ra thông tin mô hình NN
nn_model.summary()
```

```python
# Huấn luyện mô hình Neural Network
nn_model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)

# Dự đoán trên tập kiểm tra
nn_pred = nn_model.predict(X_test)
```

```python
# Tính độ chính xác trên tập kiểm tra
nn_accuracy = nn_model.evaluate(X_test, y_test)


# In ra độ chính xác của mô hình Neural Network
print('Accuracy of the NN model:', nn_accuracy)
```

```python
# Xây dựng mô hình Convolutional Neural Network
cnn_model = Sequential()
cnn_model.add(Conv1D(32, 3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(Dense(1, activation='sigmoid'))

# Compile mô hình Convolutional Neural Network
cnn_model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.001), metrics=['accuracy'])

# In ra thông tin mô hình CNN
cnn_model.summary()
```

```python
# Huấn luyện mô hình Convolutional Neural Network
cnn_model.fit(X_train_cnn, y_train, epochs=50, batch_size=32, verbose=1)

# Dự đoán trên tập kiểm tra
cnn_pred = cnn_model.predict(X_test_cnn)
```

```python
# Tính độ chính xác trên tập kiểm tra
cnn_accuracy = cnn_model.evaluate(X_test, y_test)

# In ra độ chính xác của mô hình CNN
print('Accuracy of the CNN model:', cnn_accuracy)
```
